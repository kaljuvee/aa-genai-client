```python
import pandas as pd

# 1. Load the CSV file
df = pd.read_csv('ADX_Export_APC_Tag_Values.csv')

# 2. & 3. Filter the dataframe for the specific IDX_Tag_Name value
#filtered_df = df[df['IDX_TagName'].str.contains('SIS-JIG.SEP.APC-J140_BIN_005C', case=True, na=False)]
#filtered_df = df[df['IDX_TagName'].str.contains('APC-J140_BIN_005C', case=True, na=False)]
filtered_df = df[df['IDX_TagName'].str.contains('APC-J140_BIN_005C', case=True, na=False)]

# Display the first few rows of the filtered dataframe
filtered_df.head()
```

```python
print(filtered_df.shape)
```

```python
def check_threshold(df):
    # Create masks for violations
    below_min_mask = df['ValueReal'] < df['IDX_Minimum']
    above_max_mask = df['ValueReal'] > df['IDX_Maximum']

    # Combine masks to get all violations
    violations_df = df[below_min_mask | above_max_mask].copy()

    # Print alerts for each violation
    for _, row in violations_df.iterrows():
        if row['ValueReal'] < row['IDX_Minimum']:
            print(f"{row['IDX_TagName']}: Current value {row['ValueReal']} is less than Minimum value {row['IDX_Minimum']}")
        else:
            print(f"{row['IDX_TagName']}: Current value {row['ValueReal']} is greater than Maximum value {row['IDX_Maximum']}")

    return violations_df
```

```python
df = pd.DataFrame(filtered_df)
violations = check_threshold(df)
```
